```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df1 = pd.read_parquet('/content/all_news_from_2017-08-08_to_2025-07-18.parquet')
df2 = pd.read_parquet('/content/newsday_articles_2025-07-18_to_2025-11-22.parquet')

print(df1.info())
print(df2.info())
```

```python
# Step 1: Find common columns
common_cols = df1.columns.intersection(df2.columns)

# Step 2: Keep only common columns in both DataFrames
df1_common = df1[common_cols].copy()
df2_common = df2[common_cols].copy()

# Step 3: Convert date columns to datetime if they exist
for col in ['date_iso', 'date_from_url']:
    if col in common_cols:
        df1_common[col] = pd.to_datetime(df1_common[col])
        df2_common[col] = pd.to_datetime(df2_common[col])

# Step 4: Concatenate
combined = pd.concat([df1_common, df2_common], ignore_index=True)

# Step 5: Sort by date_from_url if it exists
if 'date_from_url' in combined.columns:
    combined = combined.sort_values(by='date_from_url').reset_index(drop=True)

display(combined.head(20))
display(combined.tail(20))
display(combined.info())
```

```python
import pandas as pd

def clean_news_df(df):
    """
    Clean a news DataFrame with columns:
    ['url', 'title', 'author', 'category', 'date_iso', 'body', 'date_from_url']

    Steps:
    - Strip whitespace, fix newlines, standardize text
    - Convert dates to datetime
    - Remove duplicates and rows with missing essential columns
    """

    df = df.copy()

    # Columns to keep
    columns_to_keep = ['url', 'title', 'author', 'category', 'date_iso', 'body', 'date_from_url']
    df = df[[col for col in columns_to_keep if col in df.columns]]

    # Clean URL
    if 'url' in df.columns:
        df['url'] = df['url'].astype(str).str.strip()

    # Clean title
    if 'title' in df.columns:
        df['title'] = df['title'].astype(str)\
                        .str.replace('\n', ' ', regex=True)\
                        .str.strip()

    # Clean author
    if 'author' in df.columns:
        df['author'] = df['author'].astype(str)\
                          .str.replace('\n', ' ', regex=True)\
                          .str.strip()

    # Clean category
    if 'category' in df.columns:
        df['category'] = df['category'].astype(str)\
                                .str.strip()\
                                .str.title()

    # Clean body
    if 'body' in df.columns:
        df['body'] = df['body'].astype(str)\
                        .str.replace('\n', ' ', regex=True)\
                        .str.replace('\t', ' ', regex=True)\
                        .str.replace(' +', ' ', regex=True)\
                        .str.strip()

    # Convert dates
    for date_col in ['date_iso', 'date_from_url']:
        if date_col in df.columns:
            df[date_col] = pd.to_datetime(df[date_col], errors='coerce')

    # Remove rows with missing essential columns
    essential_cols = ['url', 'title', 'body']
    df = df.dropna(subset=[col for col in essential_cols if col in df.columns])

    # Remove duplicates by URL
    if 'url' in df.columns:
        df = df.drop_duplicates(subset='url')

    # Reset index
    df = df.reset_index(drop=True)

    return df

# usage of function

news = clean_news_df(combined)
display(news.tail(20))
```

```python
import pandas as pd

def clean_news_df(df, min_body_len=20, min_title_len=5, verbose=True):
    """
    Clean a news DataFrame with columns:
    ['url', 'title', 'author', 'category', 'date_iso', 'body', 'date_from_url']

    Steps:
    - Strip whitespace, fix newlines, standardize text
    - Convert dates to datetime
    - Remove duplicates and rows with missing essential columns
    - Remove very short articles
    - Detect '[UPDATED]' in titles
    """

    df = df.copy()
    initial_rows = len(df)

    # Columns to keep
    columns_to_keep = ['url', 'title', 'author', 'category', 'date_iso', 'body', 'date_from_url']
    df = df[[col for col in columns_to_keep if col in df.columns]]

    # Clean URL
    if 'url' in df.columns:
        df['url'] = df['url'].astype(str).str.strip()

    # Clean title
    if 'title' in df.columns:
        df['title'] = df['title'].astype(str)\
                        .str.replace('\n', ' ', regex=True)\
                        .str.strip()
        # Detect and remove [UPDATED]
        # df['is_story_updated'] = df['title'].str.contains(r'\[UPDATED\]')
        df['title'] = df['title'].str.replace(r'\[UPDATED\]', '', regex=True).str.strip()

    # Clean author
    if 'author' in df.columns:
        df['author'] = df['author'].astype(str)\
                          .str.replace('\n', ' ', regex=True)\
                          .str.strip()

    # Clean category
    if 'category' in df.columns:
        df['category'] = df['category'].astype(str)\
                                .str.strip()\
                                .str.title()

    # Clean body
    if 'body' in df.columns:
        df['body'] = df['body'].astype(str)\
                        .str.replace('\n', ' ', regex=True)\
                        .str.replace('\t', ' ', regex=True)\
                        .str.replace(' +', ' ', regex=True)\
                        .str.strip()

    # Convert dates
    for date_col in ['date_iso', 'date_from_url']:
        if date_col in df.columns:
            df[date_col] = pd.to_datetime(df[date_col], errors='coerce')

    # Remove rows with missing essential columns
    essential_cols = ['url', 'title', 'body']
    df = df.dropna(subset=[col for col in essential_cols if col in df.columns])

    # Remove empty strings after stripping
    for col in essential_cols:
        df = df[df[col].str.strip().astype(bool)]

    # Remove very short articles
    df = df[(df['body'].str.len() >= min_body_len) & (df['title'].str.len() >= min_title_len)]

    # Remove duplicates by URL
    if 'url' in df.columns:
        df = df.drop_duplicates(subset='url')

    # Reset index
    df = df.reset_index(drop=True)

    if verbose:
        final_rows = len(df)
        print(f"Initial rows: {initial_rows}, After cleaning: {final_rows}, Removed: {initial_rows - final_rows}")

    return df

# Usage
# news = clean_news_df(news)
# display(news.tail(20))
```

```python
import pandas as pd
from datetime import datetime, timedelta

# Load your full dataset
df = pd.read_parquet("/content/newsday_articles_2018-02-08_to_2025-11-22.parquet")

# Clean the DataFrame
df = clean_news_df(df)

# Ensure dates are datetime
df["date_iso"] = pd.to_datetime(df["date_iso"], errors="coerce")

# Filter for the last 1 year from today
one_year_ago = pd.Timestamp(datetime.today() - timedelta(days=365))
filtered = df[df["date_iso"] >= one_year_ago]

print(f"Length of filtered df (last 1 year): {len(filtered)}")

# Reset index
filtered.reset_index(drop=True, inplace=True)

# Save
filtered.to_parquet("news_last_1_year.parquet")

filtered.head()
```
